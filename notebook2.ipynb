```python
from ultralytics import YOLO
import cv2

# 加载预训练的 YOLOv8 模型
model = YOLO('F:\眼视光\文件\HRL组\YOLO\models\yolov8n.pt')  # yolov8n 是 YOLOv8 的 nano 版本，适合快速推理

# 读取图像
image_path = 'F:\眼视光\文件\HRL组\YOLO\images\img2.png'
image = cv2.imread(image_path)

# 进行目标检测
results = model(image)

# 可视化结果
result_image = results[0].plot()  # 获取带有检测框的图像

# 显示结果
cv2.imshow('YOLOv8 Detection', result_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# 保存结果图像
output_path = 'F:\眼视光\文件\HRL组\YOLO\images\results/img2_result.jpg'
cv2.imwrite(output_path, result_image)
```
